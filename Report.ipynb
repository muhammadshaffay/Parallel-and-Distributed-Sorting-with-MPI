```python
import matplotlib.pyplot as plt
```

```python
def plot(x_axis , y_axis , Label):
    
    plt.bar(x_axis, y_axis)
    plt.title('GFLOPS Plotting')
    
    plt.xlabel(Label)
    plt.ylabel('GFLOPS')
    
    plt.show()
```

##                                                                                             (Serial Version)

#### Question 1

A) Quick Sort

```python
x_axis = ['2^11', '2^13', '2^15', '2^17']
y_axis = [0.417, 0.038, 0.047, 0.068]
plot(x_axis, y_axis, "Quick Sort")
```

B) Merge Sort

```python
x_axis = ['2^11', '2^13', '2^15', '2^17']
y_axis = [2.127, 0.053, 0.058, 0.072]
plot(x_axis,y_axis,"Merge Sort")
```
